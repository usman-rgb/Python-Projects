```python
def condition(greater,lower):
  print(f" The Greater Number is {greater}")
  print(f" Ther Lower Number is {lower}")

User_greater=int(input(" Enter The Greater Number : "))
User_lower=int(input(" Enter The Lower Number : "))
if User_greater>User_lower:
  print(f" This Number Is Greater and The Condition is True {condition(User_greater,User_lower)}")


elif User_greater<User_lower:
  print(f" This Number Is Lower and The Condition is False {condition(User_greater,User_lower)}")

else:
  print(" Enter The Correct Number")
```

```python
def is_divisible(num, divisor):
    return num % divisor == 0

result = is_divisible(10, 2)
print(result)
```

```python
def calculate_value(price, tax_rate=4.5):
  tax_amount=price*(tax_rate/100)
  total_value=price + tax_amount
  return total_value

print(calculate_value(1000))
print(calculate_value(1000,15))
```

```python
def calculate_area(length, width):
    return length * width

length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

area = calculate_area(length, width)
print(f"The area of the rectangle is: {area}")
```

```python
def power(base, exponent):
    return base ** exponent

base = float(input("Enter the base: "))
exponent = int(input("Enter the exponent: "))

result = power(base, exponent)
print(f"The result of {base} raised to the power {exponent} is: {result}")
```
