```python
i=0
while i<20:
  print(i)
  i+=1
```

```python
sum=0
i=1
while i<5:
  sum+=i
  i+=1
  print("Sum:", sum)
```

```python
Product=2
i=1
while i<23:
  # Product*=i
  Product=Product*i
  i+=1
  print("Product:", Product)
```

```python
Division=2
i=1
num1=int(input(" Enter The Number "))
while i<num1:
  Division//=i
  i+=1
  print("Division:", Division)
```
